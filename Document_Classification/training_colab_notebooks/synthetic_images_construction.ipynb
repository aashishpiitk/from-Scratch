```python
from pathlib import Path
import cv2
```

```python
!pip install trdg
import trdg
!mkdir custom_data_32x32
!trdg -c 1000 --output_dir custom_data_32x32 -w 1 -f 32 -wd 32
checking = []
for path in Path("custom_data_32x32/").rglob("*.jpg"):
    checking.append(path)

!mkdir words_32x32
from tqdm.notebook import tqdm
for img_path in tqdm(checking):
    img = cv2.imread(str(img_path))
    _, name = str(img_path).split("/")
    for i in range(32):
        for j in range(32):
            if (img[i,j]>(150,150, 150)).all():
                img[i,j] = (255, 255, 255)
    
    cv2.imwrite(f"words_64x128/{name}", img)
```
