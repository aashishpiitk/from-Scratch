```
from PIL import Image, ImageOps
import torch
import torchvision
from torchvision.transforms import ToTensor, Scale, Compose
import matplotlib.pyplot as plt
import numpy as np
```

```
im = Image.open("lr_download.png")
gray = ImageOps.grayscale(im)
m, n = gray.size

Transform = Compose([
                     Scale(((n//32)*32,(m//32)*32)),
                     ToTensor()
])
imt = Transform(gray)
```

```
imt.shape, m, n
```

```
plt.imshow(imt[0].detach().numpy(), cmap="gray")
```

```
# Now breaking the image into 4 parts
image_index = []
for i in range(2):
    for j in range(2):
        image_patch = imt[0][i*(n//2):(i+1)*(n//2),j*(m//2):(j+1)*(m//2)]
        image_index.append(image_patch)
```
