```python
## loop to process all images in the handwritten folder
from PIL import Image, ImageOps
c=1
from pathlib import Path
from pdf2image import pdfinfo_from_path,convert_from_path

for path in Path('/content/data1/Sample Data For DocEye').rglob("*.pdf"):
    path = str(path)

    info = pdfinfo_from_path(path, userpw=None, poppler_path=None)

    maxPages = info['Pages']
    images=[]
    for page in range(1, maxPages+1, 5) : 
      images.append(convert_from_path(path, first_page=page, last_page = min(page+5-1,maxPages)))
    
   # images = convert_from_path(path) ## a list of images from the pdf
    folder_path = '/'.join(path.split('/')[:-1]) + '/'

    for image_batch in images:
      for image in image_batch:
        c+=1
        image = ImageOps.grayscale(image)
        image.save(folder_path + f'{c}.png')
        #print(folder_path + f'{c}.png')
```
